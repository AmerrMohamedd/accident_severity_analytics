```python
def classify_description(desc):
    if 'accident' in desc.lower():
        return 'Accident'
    elif 'disabled vehicle' in desc.lower():
        return 'Disabled Vehicle'
    elif 'hazard' in desc.lower():
        return 'Hazard'
    else:
        return 'Other'

df['Description'] = df['Description'].apply(classify_description)
```

```python
wind_direction_mapping = {
    'N': 'N', 'S': 'S', 'E': 'E', 'W': 'W',
    'CALM': 'No Wind', 'Variable': 'Variable', 'Unknown': 'Unknown'
}

df['Wind_Direction'] = df['Wind_Direction'].replace(wind_direction_mapping)
```

```python
import re
import pandas as pd

def clean_street_name(name):
    if pd.isna(name):
        return "Unknown"
    
    name = name.strip().lower()

    if name in ["nan", "NAN", "NaN"]:
        return pd.NA

    name = re.sub(r'^\s*[nswe]\s+|\s+[nswe]\s*$', '', name)
    
    replacements = {
        r'\bst\b': 'street',
        r'\brd\b': 'road',
        r'\bdr\b': 'drive',
        r'\bave\b': 'avenue',
        r'\bblvd\b': 'boulevard',
        r'\bhwy\b': 'highway',
        r'\bln\b': 'lane',
        r'\bct\b': 'court',
        r'\bpkwy\b': 'parkway',
        r'\bsq\b': 'square'
    }
    
    for pattern, repl in replacements.items():
        name = re.sub(pattern, repl, name)
    
    name = re.sub(r'\s+', ' ', name)

    return name.strip()

df['Street'] = df['Street'].replace([None, "", "nan", "NAN", "NaN"], pd.NA)

df['Street'] = df['Street'].str.strip()

df['Street'] = df['Street'].apply(lambda x: clean_street_name(x).strip())
```

```python
temp = df['Wind_Speed(mph)'].dropna()
Q1 = temp.quantile(0.25)
Q3 = temp.quantile(0.75)
IQR = Q3 - Q1

lower_bound_wind = Q1 - 2 * IQR
upper_bound_wind = Q3 + 2 * IQR

outliers = temp[(temp < lower_bound_wind) | (temp > upper_bound_wind)]
print(f"Number of outliers: {len(outliers)}")
```
